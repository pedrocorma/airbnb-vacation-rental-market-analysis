Conclusiones:

* beds: se puede intentar imputar a partir de accomodates.
* bedrooms sí es una variable importante para la compañía, puede intentarse imputar los nulos a través de proxies como accomodates o beds.

```python
pd.crosstab(listings_det.beds, listings_det.accommodates)
```

Parece que sí podría hacerse una imputación de nulos de la variable beds a partir de accomodates de forma más menos directa. Leyendo la matriz en vertical se tiene que:

* una o dos personas se suelen corresponder con una cama.
* tres o cuatro personas se suelen corresponder con dos camas.
* cinco o seis personas se suelen corresponder con tres camas.
* a más de 6 personas se le asignarán cuatro camas.

Número de nulos y la frecuencia de cada valor.

```python
listings_det['beds'].value_counts(dropna = False)
```

Función para imputar los nulos de beds en base a accommodates

```python
def imputar_nulos(registro):
    #Lista de condiciones
    condiciones = [(registro.accommodates <= 2),
               (registro.accommodates > 2) & (registro.accommodates <= 4),
               (registro.accommodates > 4) & (registro.accommodates <= 6),
               (registro.accommodates > 6)]

    #Lista de resultados
    resultados = [1,2,3,4]
    
    #Salida
    return(np.select(condiciones,resultados, default = -999))

#Imputación
listings_det.loc[listings_det.beds.isna(),'beds'] = listings_det.loc[listings_det.beds.isna()].apply(imputar_nulos, axis = 1).astype('float64')
```

Comprobación.

```python
listings_det.beds.value_counts(dropna = False)
```

Comprobación de la viabilidad de hacer una imputación de bedrooms.

Se comenzará cruzando el número de habitaciones con el número de personas que se pueden acomodar.

```python
pd.crosstab(listings_det.bedrooms, listings_det.accommodates)
```

No parece muy fiable.

Contrastando con el número de camas.

```python
pd.crosstab(listings_det.bedrooms, listings_det.beds, dropna=False)
```

Aquí sí podría hacerse una asignación más directa. Leyendo la matriz en vertical se observa que:

* cero, una o dos camas se suele corresponder con una habitación.
* tres o cuatro camas se suele corresponder con dos habitaciones.
* cinco o seis camas se suele corresponder con tres habitaciones.
* a más camas le vamos se le asignarán cuatro habitaciones.

Conteo de bedrooms.

```python
listings_det.bedrooms.value_counts(dropna = False)
```

```python
def imputar_nulos(registro):
    #Lista de condiciones
    condiciones = [(registro.beds <= 2),
               (registro.beds > 2) & (registro.beds <= 4),
               (registro.beds > 4) & (registro.beds <= 6),
               (registro.beds > 6)]

    #Lista de resultados
    resultados = [1,2,3,4]
    
    #Salida
    return(np.select(condiciones,resultados, default = -999))

#Imputación
listings_det.loc[listings_det.bedrooms.isna(),'bedrooms'] = listings_det.loc[listings_det.bedrooms.isna()].apply(imputar_nulos, axis = 1).astype('float64')
```

Comprobación.

```python
listings_det.bedrooms.value_counts(dropna = False)
```

Eliminando bathrooms.

```python
listings_det.drop(columns = 'bathrooms', inplace = True)
listings_det
```

## DATAMART ANALITICO

2 tablas principales:

* listings
* listings_det

Ambas comparten el campo id.

Además tambien se tiene la tabla del precio, que en este caso cruza conceptualmente con listings a través del distrito (neighbourhood_group). Nota: No se ha comprobado todavía que los literales sean iguales, por tanto quizá será necesario hacer alguna corrección manual.

```python
df = pd.merge(left = listings, right = listings_det, how = 'left', on = 'id')
```

Análisis de la variable distrito en ambas tablas, ya que necesitan coincidir para que podamos cruzarlos.

```python
distritos1 = pd.Series(df.neighbourhood_group.unique().categories).sort_values()
distritos1
```

```python
distritos2 = precio_m2.distrito
distritos2
```

Comparando parece todo igual excepto:

* Fuencarral - El Pardo
* Moncloa - Aravaca
* San Blas - Canillejas

Por tanto se remplazarán estos valores en precio_m2 para que sean iguales a los de df y puedan cruzarse.

```python
precio_m2.distrito = precio_m2.distrito.map({'Fuencarral':'Fuencarral - El Pardo',
                        'Moncloa':'Moncloa - Aravaca',
                        'San Blas':'San Blas - Canillejas'}) \
                    .fillna(precio_m2.distrito)

precio_m2
```

```python
df = pd.merge(left = df, right = precio_m2, how = 'left', left_on='neighbourhood_group', right_on='distrito')
df
```

Comprobación de que no se hayan generado nulos en la unión.

```python
df.precio_m2.isna().sum()
```

## GUARDAR EN LA BASE DE DATOS

Guardado del tablón analítico en la base de datos.

```python
df.to_sql('df', con = con, if_exists = 'replace')
```
